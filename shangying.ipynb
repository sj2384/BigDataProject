# Pie plot of offense level

```python
outputs = os.listdir('D:/Courses/DSGA1004/gg/offense_lv')
outputs = [f for f in outputs if f[-3:]=='out']
```

```python
offense_lv = ['VIOLATION','FELONY','MISDEMEANOR']
def to_list (filename,key_list):
    with open(filename, "r") as f:
        D = {}
        listname = []
        for line in f:
            key, values = line.split('\t')
            D[key] = int(values.strip('\n'))
        for k in key_list:
            listname.append(D[k])
        return listname
```

```python
to_list ('D:/Courses/DSGA1004/gg/offense_lv/offense_lv_jan.out',offense_lv)
```

```python
#month = ['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec','total']
for file in outputs:
    if file[-7:-4] == 'jan':
        jan = to_list('D:/Courses/DSGA1004/gg/offense_lv/'+file,offense_lv)
    if file[-7:-4] == 'feb':
        feb = to_list('D:/Courses/DSGA1004/gg/offense_lv/'+file,offense_lv)
    if file[-7:-4] == 'mar':
        mar = to_list('D:/Courses/DSGA1004/gg/offense_lv/'+file,offense_lv)
    if file[-7:-4] == 'apr':
        apr = to_list('D:/Courses/DSGA1004/gg/offense_lv/'+file,offense_lv)
    if file[-7:-4] == 'may':
        may = to_list('D:/Courses/DSGA1004/gg/offense_lv/'+file,offense_lv)
    if file[-7:-4] == 'jun':
        jun = to_list('D:/Courses/DSGA1004/gg/offense_lv/'+file,offense_lv)
    if file[-7:-4] == 'jul':
        jul = to_list('D:/Courses/DSGA1004/gg/offense_lv/'+file,offense_lv)
    if file[-7:-4] == 'aug':
        aug = to_list('D:/Courses/DSGA1004/gg/offense_lv/'+file,offense_lv)
    if file[-7:-4] == 'sep':
        sep = to_list('D:/Courses/DSGA1004/gg/offense_lv/'+file,offense_lv)
    if file[-7:-4] == 'oct':
        Oct = to_list('D:/Courses/DSGA1004/gg/offense_lv/'+file,offense_lv)
    if file[-7:-4] == 'nov':
        nov = to_list('D:/Courses/DSGA1004/gg/offense_lv/'+file,offense_lv)
    if file[-7:-4] == 'dec':
        dec = to_list('D:/Courses/DSGA1004/gg/offense_lv/'+file,offense_lv)
    else:
        total = to_list('D:/Courses/DSGA1004/gg/offense_lv/'+file,offense_lv)
    
    
```

```python
def plot_pie(monlist, month):
    colors = ['gold', 'yellowgreen', 'lightskyblue']
    plt.pie(monlist, labels=offense_lv, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
 
    plt.axis('equal')
    plt.title('offense of ' + month)
    plt.savefig('D:/Courses/DSGA1004/gg/offense_lv/'+month+'.png', format='png')
    plt.show()
```

```python
for x,y in zip(monlist, month):
    plot_pie(x,y)
```

```python
def to_dict(filename):
    with open(filename, "r") as f:
        D = {}
        for line in f:
            key, values = line.split('\t')
            D[key] = int(values.strip('\n'))
        return D
```
